### 任务四：双向链表的实现（难度：★★）

**实现基本功能**：（包括但不限于，可以根据自己能力继续扩展）

　　（1）同单链表一样（10个函数均要实现）

　　（2）反转链表：通过操控指针的方式依次倒序输出链表值；
	
　【注】：单链表的结点类只有一个指针：**next指针**，而双向链表的结点类有两个指针，分别为**prior指针**和**next指针**，prior指针指向前一个结点，next指向后一个结点！

**最后输出结果要求**：
