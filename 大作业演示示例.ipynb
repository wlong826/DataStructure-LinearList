### 关键点讲解：

- 这里只简单举一个尾插法的例子说明下单向循环链表

```python
# 尾插法
def append(self, newItem):
    node = SingleNode(newItem)
    nod = self.head # 定义nod为临时结点
    while nod.next != self.head: # 当nod下一结点为头结点时，停止遍历
        nod = nod.next
    node.next = nod.next # 在语句执行之前，尾结点指向头结点，所以先让新结点的next指向尾结点的next(即为头结点)
    nod.next = node
```
