### 1. 定义类

```python
import random
random.seed(10)

# 定义结点类，包含两个成员：结点元素值和指向下一结点的指针
class SingleNode():
    # 结点类初始化
    def __init__(self, item):
        self.item = item  # item存放结点的数值
        self.next = None

# 定义单链表，
class SingleCycleLinkList():
    # 链表类初始化
    def __init__(self):
        self.head = None

    # 判断链表是否为空
    def is_empty(self):
        return self.head == None
    
    # 遍历整个链表
    def travel(self):
        show_list = []
        if self.is_empty():
            print("链表为空！")
            return None
        nod = self.head
        while nod.next != self.head: # 当结点的下一个指向不是头结点时
            show_list.append(nod.item)
            nod = nod.next
        show_list.append(nod.item) # 将尾结点的item添加到列表里
        return show_list
    
    # 头插法创建单链表
    def add(self, newItem):
        node = SingleNode(newItem) # 定义新结点对象
         # 如果为空链表
        if self.is_empty():
            self.head = node # 则头结点即为当前结点
            node.next = node # 只有一个头结点时，那么头结点下一指向即为当前结点（单向循环）
            return # 跳出函数
        # 如果不是空链表，那么就遍历找到尾结点
        nod = self.head
        while nod.next != self.head: # 当前结点的next为头结点时，停止遍历
            nod = nod.next
        node.next = nod.next # 在语句执行之前，尾结点指向头结点，所以先让新结点的next指向尾结点的next(即为头结点)
        nod.next = node # 再让尾结点的next指向新结点
        self.head = node # 头结点即为插入的新结点

    # 尾插法
    def append(self, newItem):
        node = SingleNode(newItem)
        nod = self.head # 定义nod为临时结点
        while nod.next != self.head: # 当nod下一结点为头结点时，停止遍历
            nod = nod.next
        node.next = nod.next # 在语句执行之前，尾结点指向头结点，所以先让新结点的next指向尾结点的next(即为头结点)
        nod.next = node
    
    # 输出链表长度
    def get_length(self):
        return len(self.travel())
    
    # 指定位置添加元素
    def insert(self, pos, newItem):
        node = SingleNode(newItem)
        if pos < 1 or pos > self.get_length():
            print("输入的索引有误！")
            return False
        elif pos == 1: 
            self.add(newItem)
        elif pos == self.get_length():
            self.append(newItem)
        else:
            nod = self.head
            j = 1
            while j < pos - 1:
                nod = nod.next
                j += 1
            node.next = nod.next
            nod.next = node
        return True
    
    # 删除指定位置上的结点
    def remove(self, pos):
        if pos < 1 or pos > self.get_length():
            print("输入的索引有误！")
            return False
        else:
            nod = self.head
            j = 1
            while j < pos - 1:
                nod = nod.next
                j += 1
            nod.next = nod.next.next
            return True

    # 查找指定位置的结点值
    def find(self, pos):
        if pos < 1 or pos > self.get_length():
            print("输入的索引有误！")
            return False
        else:
            nod = self.head
            j = 1
            while j < pos: # 找到指定位置得结点，停止循环
                nod = nod.next
                j += 1
            return nod.item
        
    # 更新链表中某个位置的值
    def updata(self, pos, newItem):
        if pos < 1 or pos > self.get_length():
            print("输入的索引有误！")
            return False
        else:
            nod = self.head
            j = 1
            while j < pos: # 找到指定位置得结点，停止循环
                nod = nod.next
                j += 1
            nod.item = newItem
            return True

    # 清空链表
    def clear(self):
        self.head = None
    
```

### 2. 创建单链表并进行遍历

#### 2.1 头插法

```python
length = 4 # 定义链表长度
linklist = SingleCycleLinkList()
newItems = [random.randint(1,100) for i in range(length)]
print(newItems)
for newItem in newItems:
    linklist.add(newItem) # 头插法创建单链表
linklist.travel() # 遍历单链表
```

#### 2.2 尾插法

```python
print(linklist.travel()) # 尾插法前的单链表
for newItem in newItems:
    linklist.append(newItem) # 尾插法往单链表添加元素
linklist.travel() # 尾插法后的单链表
```

### 3. 输出单链表长度

```python
len_ = linklist.get_length()
print("当前链表长度为{}".format(len_))
```

### 4. 指定位置添加元素

```python
pos = 3
newItem = 100
if linklist.insert(pos, newItem): # 如果插入成功，则遍历链表
    print(linklist.travel())
```

### 5. 删除指定位置上的值

```python
pos = 5 # 要删除值得指定位置
if linklist.remove(pos):
    print(linklist.travel())
```

### 6. 查找指定位置上的值

```python
pos = 4
if linklist.find(pos):
    print("当前链表中索引为{}的值为{}".format(pos, linklist.find(pos)))
```

### 7. 更新链表中某个位置的值

```python
pos = 6
newItem = 99
if linklist.updata(pos, newItem):
    print(linklist.travel())
```

### 8. 清空链表

```python
linklist.clear() # 清空链表
linklist.travel() # 遍历链表
```
