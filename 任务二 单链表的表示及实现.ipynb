### 1. 定义结点类和单链表类

```python
import random
random.seed(10)

# 定义结点类，包含两个成员：结点元素值和指向下一结点的指针
class SingleNode():
    # 结点类初始化
    def __init__(self, item):
        self.item = item  # item存放结点的数值
        self.next = None

# 定义单链表，
class SingleLinkList():
    
    # 链表类初始化
    def __init__(self):
        self.head = None
    
    # 判断链表是否为空
    def is_empty(self):
        return self.head == None
    
    # 输出链表长度
    def get_length(self):
        return len(self.travel())
    
    # 遍历整个链表
    def travel(self):
        show_list = [] # 创建个空数组，存储链表各结点元素值
        if self.is_empty():
            return None
        nod = self.head # 从头节点开始遍历
        while nod != None: # 当结点为空时，到链表尾部，停止遍历
            show_list.append(nod.item)
            nod = nod.next # 跳向下一个结点
        return show_list
    
    # 头插法创建单链表
    def add(self, newItem):
        node = SingleNode(newItem)
        node.next = self.head # 指针变换
        self.head = node
    
    # 尾插法
    def append(self, newItem):
        node = SingleNode(newItem)
        # 从头结点开始遍历
        nod = self.head
        while nod.next != None: # 当下一个结点的next为None时，停止遍历//  注：跟网上不一样，回头用尾插法新建单链表试试
            nod = nod.next
        nod.next = node
        
     # 指定位置添加元素
    def insert(self, pos, newItem): # 在指定i位置上添加e元素
        if pos < 1 or pos > self.get_length():
            print("输入的位置有误！")
            return False
        elif pos == 1:
            self.add()
        elif pos == self.get_length():
            self.append()
        else: # 当要添加的元素位置，既不在链表头部也不再尾部时
            j = 1 # 当j增加到i时
            add_nod = SingleNode(newItem) # 要添加的结点
            nod = self.head # 临时结点变量
            while j < pos - 1:
                nod = nod.next
                j += 1
            add_nod.next = nod.next 
            nod.next = add_nod
            return True
            
    # 删除指定位置上的结点
    def remove(self, pos):
        if pos < 1 or pos > self.get_length():
            print("输入的位置有误！")
            return False
        else:
            j = 1
            nod = self.head
            while j < pos-1:
                nod = nod.next
                j += 1
            nod.next = nod.next.next
            return True

    # 查找指定位置的结点值
    def find(self, pos):
        if pos < 1 or pos > self.get_length():
            print("输入的位置有误！")
            return False
        else:
            j = 1
            nod = self.head
            while j < pos:
                nod = nod.next
                j += 1
            return nod.item
    # 更新链表中某个位置的值
    def update(self, pos, newItem):
        if pos < 1 or pos > self.get_length():
            print("输入的位置有误！")
            return False
        else:
            j = 1
            nod = self.head
            while j < pos:
                nod = nod.next
                j += 1
            nod.item = newItem
            return True
    
    # 清空链表
    def clear(self):
        self.head = None
```

### 2. 创建单链表并进行遍历

#### 2.1 头插法

```python
length = 4 # 定义链表长度
linklist = SingleLinkList()
newItems = [random.randint(1,100) for i in range(length)]
print(newItems)
for newItem in newItems:
    linklist.add(newItem) # 头插法创建单链表
linklist.travel() # 遍历单链表
```

#### 2.2 尾插法

```python
for newItem in newItems:
    linklist.append(newItem) # 尾插法创建单链表
linklist.travel() # 遍历单链表
```

### 4. 指定位置添加元素

```python
pos = 5  # 指定位置
newItem = 100 # 结点元素值

# 向指定位置添加新的结点
if linklist.insert(pos, newItem): # 判断下是否插入成功，成功则遍历链表
    print(linklist.travel()) # 遍历单链表
```

### 5. 删除指定位置上的值

```python
pos = 5 # 指定要删除结点的位置
if linklist.remove(pos):
    print(linklist.travel())
```

### 6. 查找指定位置上的值

```python
pos = 7 # 指定位置
if linklist.find(pos):
    print(linklist.find(pos))
```

```python
pos = 6
newItem = 10
if linklist.update(pos, newItem):
    print(linklist.travel())
```

```python
linklist.clear()
print(linklist.travel())
```
