### 1. 定义结点类和双向链表类:DoubleLinklist()

```python
import random
random.seed(10)

# 定义结点类，包含三个成员
class SingleNode():
    # 结点类初始化
    def __init__(self, item):
        self.item = item  # item存放结点的数值
        self.prior = None # 指向上一结点的指针
        self.next = None # 指向下一结点的指针

# 定义双向链表
class DoubleLinkList():
    # 链表类初始化
    def __init__(self):
        self.head = None # 头结点
        self.tail = None # 尾结点
    
    # 判断链表是否为空
    def is_empty(self):
        return self.head == None
    
    # 输出链表长度
    def get_length(self):
        return len(self.travel())
    
    # 遍历整个链表
    def travel(self):
        show_list = [] # 创建个空数组，存储链表各结点元素值
        if self.is_empty():
            print("链表为空！")
            return None
        nod = self.head # 从头节点开始遍历
        while nod != None: # 当结点为空时，到链表尾部，停止遍历
            show_list.append(nod.item)
            nod = nod.next # 跳向下一个结点
        return show_list
    
    # 头插法创建单链表
    def add(self, newItem):
        node = SingleNode(newItem)
        # 如果为空
        if self.is_empty():
            node.next = self.head # 指针变换
            self.tail = node # 当只有一个结点时，头结点即为尾结点，且头插法只定义这一次尾结点即可，因为是往前添加结点
            self.head = node # 头结点即为当前结点
            return
        # 如果不为空
        node.next = self.head
        self.head.prior = node
        self.head = node
    
    # 尾插法
    def append(self, newItem):
        node = SingleNode(newItem)
        if self.is_empty(): # 链表为空时，头结点与尾结点均为新插入的结点
            self.head = node 
            self.tail = node
            return
        # 从头结点开始遍历
        nod = self.head
        while nod.next != None: # 当下一个结点的next为None时，停止遍历//  注：跟网上不一样，回头用尾插法新建单链表试试
            nod = nod.next
        nod.next = node
        node.prior = nod
        self.tail = node # 尾插法时尾结点即为新插入的结点
    
    # 链表反转
    def reverse(self):
        reverse_list = [] # 创建个空数组，存储链表反转后各结点元素值
        if self.is_empty():
            print("链表为空！")
            return None
        nod = self.tail # 从尾节点开始遍历
        while nod != self.head: # 从尾结点使用前向指针往前遍历
            reverse_list.append(nod.item)
            nod = nod.prior # 前进一个结点
        reverse_list.append(nod.item) # 加入头结点的值
        return reverse_list
    
    # 指定位置添加元素
    def insert(self, pos, newItem): # 在指定pos位置上添加数据为newItem的结点
        if pos < 1 or pos > self.get_length():
            print("输入的位置有误！")
            return False
        elif pos == 1:
            self.add()
        elif pos == self.get_length():
            self.append()
        else: # 当要添加的元素位置，既不在链表头部也不再尾部时
            j = 1 # 当j增加到i时
            add_nod = SingleNode(newItem) # 要添加的结点
            nod = self.head # 临时结点变量
            while j < pos - 1:
                nod = nod.next
                j += 1
            add_nod.next = nod.next 
            nod.next = add_nod
            return True
    
    # 删除指定位置上的结点
    def remove(self, pos):
        if pos < 1 or pos > self.get_length():
            print("输入的位置有误！")
            return False
        else:
            j = 1
            nod = self.head
            while j < pos-1:
                nod = nod.next
                j += 1
            nod.next = nod.next.next
            return True
    
     # 查找指定位置的结点值
    def find(self, pos):
        if pos < 1 or pos > self.get_length():
            print("输入的位置有误！")
            return False
        else:
            j = 1
            nod = self.head
            while j < pos:
                nod = nod.next
                j += 1
            return nod.item
          
    # 更新链表中某个位置的值
    def update(self, pos, newItem):
        if pos < 1 or pos > self.get_length():
            print("输入的位置有误！")
            return False
        else:
            j = 1
            nod = self.head
            while j < pos:
                nod = nod.next
                j += 1
            nod.item = newItem
            return True
    
    # 清空链表
    def clear(self):
        self.head = None
        self.tail = None
```

### 2. 创建单链表并进行遍历

#### 2.1 头插法

```python
length = 4 # 定义链表长度
linklist = DoubleLinkList()
newItems = [random.randint(1,100) for i in range(length)]
print(newItems)
for newItem in newItems:
    linklist.add(newItem) # 头插法创建单链表
linklist.travel() # 遍历单链表
```

#### 2.2 尾插法

```python
print(linklist.travel()) # 尾插法前的单链表
for newItem in newItems:
    linklist.append(newItem) # 尾插法往单链表添加元素
linklist.travel() # 尾插法后的单链表
```

### 3. 反转链表

```python
linklist.reverse() # 链表反转
```

### 4. 输出单链表长度

```python
len_ = linklist.get_length()
print("当前链表长度为{}".format(len_))
```

### 5. 指定位置添加元素

```python
pos = 3
newItem = 100
if linklist.insert(pos, newItem): # 如果插入成功，则遍历链表
    print(linklist.travel())
```

### 6. 删除指定位置上的值

```python
pos = 5 # 要删除值得指定位置
if linklist.remove(pos):
    print(linklist.travel())
```

### 7. 查找指定位置上的值

```python
pos = 4
if linklist.find(pos):
    print("当前链表中索引为{}的值为{}".format(pos, linklist.find(pos)))
```

### 8. 更新链表中某个位置的值

```python
pos = 6
newItem = 99
if linklist.update(pos, newItem):
    print(linklist.travel())
```

### 9. 清空链表

```python
linklist.clear() # 清空链表
linklist.travel() # 遍历链表
```
