**实现基本功能**：（包括但不限于，可以根据自己能力继续扩展）

　　（1）顺序表**初始化**：
　　　　init(self, length) 
	
　　（2）在指定位置上**插入**元素：
　　　　ListInsert(self, i, e) 在i索引位置上插入元素e
	
　　（3）**遍历**顺序表中的所有元素
　　　　ShowList(self)
	
　　（4）**删除**指定位置上的元素
　　　　ListDelete(self, i)

　　（5）**取出**指定位置上的元素值
　　　　GetElem(self, i)
	
　　（6）**查找**给定元素值在顺序表中的索引位置
　　　　LocateElem(self, e)
	
**最后输出结果要求**：

### 1. 定义顺序表类

```python
import random
random.seed(10)

# 定义顺序表的类
class SqList():
    # 初始化
    def __init__(self, length):
        self.MAXSIZE = 100
        self.length = length
        self.sqlist = [random.randint(1,100) for j in range(length)] # 生成length长度的随机数组
    # 输出所有元素
    def ShowList(self):
        return self.sqlist
    # 取值
    def GetElem(self, i):
        # 首先判断插入的位置是否在合法范围内(1,length)
        if i < 0 or i > self.length: 
            print("输入的位置索引错误！")
            return False
        return self.sqlist[i-1]
    # 查找
    def LocateElem(self, e):
        for j in range(self.length):
            if self.sqlist[j] == e:
                return j
        print("列表中不存在您查找的元素！")
        return False
    # 插入
    def ListInsert(self, i, e):
        if i < 0 or i > self.length: 
            print("输入的位置索引错误！")
            return False
        return self.sqlist.insert(i,e)
    # 删除
    def ListDelete(self, i):
        if i < 0 or i > self.length: 
            print("输入的位置索引错误！")
            return False
        return self.sqlist.pop(i)
```

### 2. 创建对象

```python
# 定义一个顺序表对象
length = 10
my_sqlist = SqList(length)
```

### 3. 遍历顺序表

```python
# 输出所有参数
mylist = my_sqlist.ShowList()
print(mylist)
```

### 4. 取值

```python
# 取值
index = 10
value = my_sqlist.GetElem(index)
if value:
    print("列表中第{}个数等于{}".format(index,value))
```

### 5. 查找

```python
# 查找
e = 55 # 要查找的数
index = my_sqlist.LocateElem(e)
if index:
    print("元素{}在列表中的索引为{}".format(e, index))
```

### 6. 插入

```python
# 插入
print("插入前的列表：{}".format(my_sqlist.ShowList()))
i = 4 # 插入的索引位置
e = 10 # 插入的值
my_sqlist.ListInsert(i, e) # 在指定索引位置插入值
print("插入后的列表：{}".format(my_sqlist.ShowList()))
```

### 7. 删除

```python
# 删除
i = 5 # 删除的索引位置
print("删除前的列表：{}".format(my_sqlist.ShowList()))
my_sqlist.ListDelete(i) # 删除i索引所在位置上的值
print("删除后的列表：{}".format(my_sqlist.ShowList()))
```
